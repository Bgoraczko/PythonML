```python
import matplotlib.pyplot as plt
Age = data['Age']
AW = data['Actual Weight']
plt.bar(Age,AW)
plt.xlabel('Age')
plt.ylabel('Actual weight')
plt.title('Bar graph - Age vs Actual weight')
```

```python
import matplotlib.pyplot as plt
DW = data['Dream Weight']
AW = data['Actual Weight']
area = 1
plt.scatter(DW,AW,area)
plt.xlabel('Dream Weight')
plt.ylabel('Actual Weight')
plt.title('Scatterplot - Dream Weight vs Actual Weight')
plt.show()
```

```python
from sklearn.model_selection import train_test_split
X = data.drop('Exercise Intensity', axis = 1)
y = data['Exercise Intensity']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2,random_state=1)
X_train, X_validation, y_train, y_validation = train_test_split(X_train, y_train, test_size=0.25,random_state=1)

print(f'Percentage ratio of the training set:{len(X_train)/len(X)*100}')
print(f'Percentage ratio of the validation set:{len(X_validation)/len(X)*100}')
print(f'Percentage ratio of the testing set:{len(X_test)/len(X)*100}')
```
