TASK 4

```python
quality = data['quality']
data = data.drop(columns = ['quality'])
from sklearn.preprocessing import (
MinMaxScaler,
StandardScaler,
)
standard_scaler = StandardScaler().set_output(transform="pandas")
data_standardized = standard_scaler.fit_transform(data)
minmax_scaler = MinMaxScaler().set_output(transform="pandas")
data_normalized = minmax_scaler.fit_transform(data)
```

```python
data_normalized.hist(figsize=(15,15))
```

```python
data_standardized.hist(figsize=(15,15))
```

TASK 6

```python
from sklearn.decomposition import PCA

pca = PCA().set_output(transform="pandas").fit(data)
data_pca = pca.transform(data)

sns.scatterplot(data=data_pca, x=data_pca['pca0'], y=data_pca['pca1'])
```

```python
sns.scatterplot(data=data_pca, x=data_pca['pca9'], y=data_pca['pca10'])
```

```python
sns.scatterplot(data=data_pca, x=data_pca['pca4'], y=data_pca['pca6'])
```

TASK 7

```python
pca1 = PCA().set_output(transform="pandas").fit(data_normalized)
data_pca1 = pca1.transform(data_normalized)

sns.scatterplot(data=data_pca1, x=data_pca1['pca0'], y=data_pca1['pca1'])
```

```python
pca2 = PCA().set_output(transform="pandas").fit(data_standardized)
data_pca2 = pca2.transform(data_normalized)

sns.scatterplot(data=data_pca2, x=data_pca2['pca0'], y=data_pca2['pca1'])
```
